```python
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as plt 
```

```python
filename = "/scratch/ddr8143/multiqa/baseline_runs_v0/ambigqa_bm25_100.from_nq.bs_48.ws_4.t_0.s_0/reader_training/26581335_reader__ambigqa_bm25_100.from_nq.bs_48.ws_4.t_0.s_0.out"
```

```python
results = {}
with open(filename) as f:
    i = 0
    e = 0
    for l in f:
        curr_EM = 0
        if "EM" in l:
            curr_EM += float(l.split()[-1])
            i += 1


        if i == 4:
            i = 0
            results[e] = curr_EM
            e += 1
```

```python
results
```

```python
vals = list(results.items())
plt.figure(figsize=(10, 5))
plt.plot([v[0] for v in vals], [v[1] for v in vals], label="Val Exact Match")
plt.xlabel("Epoch")
plt.ylabel("Percent Exact Match")
plt.title("AmbigQA Reader Training")
plt.ylim([20, 45])
```
